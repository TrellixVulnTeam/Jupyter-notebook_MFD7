```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        
    def append(self, data):
        if not self.head:
            self.head = DNode(data)
        else:
            current = self.head
            while current.next:
                current = current.next
            new_node = DNode(data)
            current.next = new_node
            new_node.prev = current
            
    def insert(self, index, value):
        if not self.head:
            self.head = DNode(data)
        elif index == 0:
            new_node = DNode(data)
            temp = self.head
            new_node.next = temp
            temp.prev = new_node
            self.head = new_node
        else:
            new_node = DNode(data)
            current = self.head
            counter = 0
            while current.next and counter != index:
                current = current.next
                counter += 1
            if current.next:
                temp = current.next
                current.next = new_node
                new_node.next = temp
                new_node.prev = current
                temp.prev = new_node
            else:
                current.next = new_node
                new_node.prev = current
                
    
    def __repr__(self):
        out_str = ""
        current = self.head
        while current:
            out_str  += "%s, " % current.data
            current = current.next
        return "[%s]" % out_str
             
    def remove(self, value):
        found = False
        if not self.head:
            raise ValueError(
                "list.remove(%s: %s is not in the list"
                % (value, value))
        else:
            if self.head.data == value:   # case 1
                self.head = self.head.next
                found = True
            else:
                current = self.head.next # cases 2 and 3
                prev = self.head
                # while current is not None and found is not False:
                while current and not found:
                    if current.data == value:
                        prev.next = current.next
                        found = True
                    prev = current
                    current = current.next
                if not found:
                     raise ValueError(
                "list.remove(%s: %s is not in the list"
                % (value, value))
                        
            
            
            
```
